```python
# Import required libraries for Image Processing

import cv2                  # OpenCV for image processing operations
import numpy as np          # NumPy for numerical operations
import matplotlib.pyplot as plt  # Matplotlib for displaying images
```

```python
# 1. Read and display the original image

# Read the image using OpenCV
image = cv2.imread('input.jpg')

# Convert BGR (OpenCV format) to RGB for correct color display
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

# Display the original image
plt.imshow(image_rgb)
plt.title("Original Image")
plt.axis('off')
```

```python
# 2. Grayscale Conversion

# Convert the RGB image to Grayscale
gray_image = cv2.cvtColor(image_rgb, cv2.COLOR_RGB2GRAY)

# Display the grayscale image
plt.imshow(gray_image, cmap='gray')
plt.title("Grayscale Image")
plt.axis('off')
```

```python
# 3. Image Resizing

# Resize the grayscale image to a fixed size (e.g., 256 x 256)
resized_image = cv2.resize(gray_image, (256, 256))

# Display the resized image
plt.imshow(resized_image, cmap='gray')
plt.title("Resized Image (256 x 256)")
plt.axis('off')
plt.show()
```

```python
# 4. Gaussian Blur (Noise Reduction)

# Apply Gaussian Blur to the resized image
blurred_image = cv2.GaussianBlur(resized_image, (5, 5), 0)

# Display the blurred image
plt.imshow(blurred_image, cmap='gray')
plt.title("Gaussian Blurred Image")
plt.axis('off')
plt.show()
```

```python
# 5. Edge Detection using Canny

# Apply Canny edge detection
edges = cv2.Canny(blurred_image, threshold1=100, threshold2=200)

# Display the edge-detected image
plt.imshow(edges, cmap='gray')
plt.title("Edge Detected Image (Canny)")
plt.axis('off')
plt.show()
```

```python
# 6. Image Thresholding

# Apply binary thresholding
_, thresholded_image = cv2.threshold(
    blurred_image, 127, 255, cv2.THRESH_BINARY
)

# Display the thresholded image
plt.imshow(thresholded_image, cmap='gray')
plt.title("Thresholded Image")
plt.axis('off')
plt.show()
```

```python
# 7. Histogram Equalization

# Apply histogram equalization on the grayscale image
equalized_image = cv2.equalizeHist(resized_image)

# Display the histogram equalized image
plt.imshow(equalized_image, cmap='gray')
plt.title("Histogram Equalized Image")
plt.axis('off')
plt.show()
```
